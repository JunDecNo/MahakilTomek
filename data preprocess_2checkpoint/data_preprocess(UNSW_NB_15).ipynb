```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

### 读取文件

```python
path = 'E:/Source/ThsisSource/DataSource/UNSW-NB15/all.csv'
data = pd.read_csv(path)
data
```

### 删除无意义的列数据

```python
Label = data['Label']
data = data.drop(['srcip', 'sport', 'dsport', 'dstip', 'Ltime', 'Stime', 'attack_cat','Label'], axis=1)
```

### 对 proto、state、service进行独热编码

```python
enc = OneHotEncoder(sparse=False)
proto = enc.fit_transform(data[['proto']])  # proto
state = enc.fit_transform(data[['state']])  # state
service = enc.fit_transform(data[['service']])  # service
data = data.drop(['proto', 'state', 'service'], axis=1)
data = np.array(data)
data = np.concatenate((data, proto, state, service), axis=1)
data = pd.DataFrame(data)
data
```
